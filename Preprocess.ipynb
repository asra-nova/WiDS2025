```python
import re
import numpy as np
import pandas as pd
```

```python
# Load CSV file
df = pd.read_csv(
    "../Data/widsdatathon2025/TRAIN/TRAIN_FUNCTIONAL_CONNECTOME_MATRICES.csv"
)  # Adjust delimiter if needed
```

```python
matrix_positions = []
for col in df.columns[1:]:  # Skip participant_id
    match = re.match(r"(\d+)throw_(\d+)thcolumn", col)
    if match:
        row_idx, col_idx = map(int, match.groups())
        matrix_positions.append((row_idx, col_idx, col))

# Determine matrix shape
n_rows = max(pos[0] for pos in matrix_positions) + 1
n_columns = max(pos[1] for pos in matrix_positions) + 1
```

```python
for index, row in df.iterrows():
    participant = row["participant_id"]
    matrix = np.zeros((n_rows, n_columns))  # Initialize empty matrix

    # Fill the matrix using the extracted column indices
    for row_idx, col_idx, col_name in matrix_positions:
        matrix[row_idx, col_idx] = row[col_name]

    print(f"Matrix for {participant}:\n", matrix.shape)
    break
    print()
```
