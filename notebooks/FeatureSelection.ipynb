```python
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
train_x = pd.read_csv('../../Data/preprocessed/train/aux.csv')
train_y = pd.read_csv('../../Data/preprocessed/train/labels.csv')
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(train_x.drop(columns='participant_id'), train_y['Sex_F'])
```

```python
coefficients = pd.Series(model.coef_[0], index=train_x.drop(columns='participant_id').columns)
```

```python
top_features = coefficients.abs().nlargest(30)
```

```python
top_features
```

```python
plt.figure(figsize=(10,6))
top_features.sort_values().plot(kind='barh', color='skyblue')
plt.title('Top 10 Features for Sex Outcome')
plt.ylabel('Features')
plt.xlabel('Absolute Coefficient Value')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(train_x.drop(columns='participant_id'), train_y['ADHD_Outcome'])
```

```python
coefficients = pd.Series(model.coef_[0], index=train_x.drop(columns='participant_id').columns)
```

```python
top_features = coefficients.abs().nlargest(30)
print(top_features)
```

```python
plt.figure(figsize=(10,6))
top_features.sort_values().plot(kind='barh', color='skyblue')
plt.title('Top 10 Features for ADHD Outcome')
plt.ylabel('Features')
plt.xlabel('Absolute Coefficient Value')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
model = LogisticRegression(penalty='l1', solver='liblinear')
model.fit(train_x.drop(columns='participant_id'), train_y['Sex_F'])
```

```python
selected_features_Sex = train_x.drop(columns='participant_id').columns[model.coef_[0] != 0]
print(selected_features_Sex)
```

```python
model = LogisticRegression(penalty='l1', solver='liblinear')
model.fit(train_x.drop(columns='participant_id'), train_y['ADHD_Outcome'])
```

```python
selected_features_ADHD = train_x.drop(columns='participant_id').columns[model.coef_[0] != 0]
print(selected_features_ADHD)
```

```python
union_features = list(set(selected_features_ADHD) | set(selected_features_Sex))
pprint(union_features)
print(len(union_features))
```

```python
print(len(train_x.columns))
```

```python
train_x_selected = train_x[['participant_id'] + union_features]
train_x_selected.to_csv('../../Data/preprocessed_selected_features/train/aux.csv', index=False)
```
