# function

```python
def normalize_df(df):
    # Extract numeric columns (excluding the first column)
    numeric_cols = df.columns[1:]
    x = df[numeric_cols].to_numpy()

    # Min-max normalization preserving sign
    abs_values = np.abs(x)
    min_ = np.min(abs_values, axis=0, keepdims=True)
    max_ = np.max(abs_values, axis=0, keepdims=True)
    abs_values = (abs_values - min_) / (max_ - min_)

    # Modify DataFrame in place
    df.loc[:, numeric_cols] = abs_values * np.sign(x)

    return df  # Returns modified DataFrame
```

```python
def normalize_two_dfs(df1, df2):
    # Extract numeric columns (excluding the first column)
    numeric_cols = df1.columns[1:]  # Assuming both DataFrames have the same structure

    # Stack both DataFrames' numeric values for global normalization
    combined_x = np.vstack([df1[numeric_cols].to_numpy(), df2[numeric_cols].to_numpy()])

    # Compute global min and max
    abs_values = np.abs(combined_x)
    min_ = np.min(abs_values, axis=0, keepdims=True)
    max_ = np.max(abs_values, axis=0, keepdims=True)
    
    # Perform min-max normalization while preserving sign
    normalized_values = (abs_values - min_) / (max_ - min_) * np.sign(combined_x)

    # Split back into two DataFrames
    df1.loc[:, numeric_cols] = normalized_values[:len(df1)]
    df2.loc[:, numeric_cols] = normalized_values[len(df1):]

    return df1, df2  # Returns both normalized DataFrames
```

```python
train, test = pd.read_csv(
    "../../Data/new/widsdatathon2025/TRAIN/TRAIN_FUNCTIONAL_CONNECTOME_MATRICES_new_36p_Pearson.csv"
), pd.read_csv("../../Data/new/widsdatathon2025/TEST/TEST_FUNCTIONAL_CONNECTOME_MATRICES.csv")
```

```python
normed_train, normed_test = normalize_two_dfs(train, test)
```
